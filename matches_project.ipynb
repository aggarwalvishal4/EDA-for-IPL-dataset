### 3.3 Preprocessing

- Dealing with missing values<br/>
    - Dropping missing entries of __umpire3.__
    - Replacing missing values of __winner__ with __N/A__ values as these matches were having __no result__.
    - Changing __city__ to _Abu Dhabi_ as they are from _Dubai_ Venue. 
    - Replacing values of __winner__ with _N/A_ values as those matches were _tie_.
    - Added a new column __loser__ to the dataset.

```python
matches[matches['winner'].isnull()] #analyzing data where winner is null
```

```python
matches[matches['city'].isnull()]   #analyzing data where city is null
```

```python
matches[matches['umpire1'].isnull()]   #analyzing data where umpire1 is null
```

```python
matches[matches['player_of_match'].isnull()] #analyzing data where man of the match is null
```

```python
matches[matches['result'] == "tie"]  #analyzing data where result is tie
```

```python
#getting index of cities where it is null
matches_processed = matches
#setting city as Abu Dhabi where city is null because venue is from Abu Dhabi
matches_processed[['city']] = matches_processed[['city']].fillna(value="Abu Dhabi")
```

```python
matches_processed[matches_processed["city"]=="Abu Dhabi"]
```

```python
np.unique(matches_processed['team1']) #checking team names
# replacing Rising Pune Supergiants with Rising Pune Supergiant as these are same teams.
matches_processed = matches_processed.where(matches_processed != 'Rising Pune Supergiants', 'Rising Pune Supergiant')
np.unique(matches_processed['team1']) #validating replacement
```

```python
matches_processed = matches_processed.drop('umpire3', axis=1) #dropping of umpire3 column
matches_processed
```

```python
matches_processed.loc[matches_processed['result'] == 'tie','winner'] = np.nan #replacing winner with NaN where match is tie
matches_processed[matches_processed['result'] == 'tie']
```

```python
matches_processed=matches_processed.replace(np.NaN,'N/A') #replacing NaN with N/A in player of match column
```

```python
matches_processed[matches_processed['player_of_match']=='N/A'] #validating NaN replacement
```

```python
#Adding of Loser Column
loser = []
for index, row in matches_processed.iterrows():
    if(row['team1'] == row['winner']):
        loser.append(row['team2'])
    elif(row['team2'] == row['winner']):
        loser.append(row['team1'])
    else:
        loser.append('N/A')
matches_processed['loser'] = pd.Series(loser, index = matches_processed.index)
matches_processed
```

### 4.13 Which team lost the toss but won the match most?

```python
toss_loss_count = {}
Teams = np.unique(matches_processed['team1'])
for i in range(0, len(Teams)):
    toss_loss_count[Teams[i]] = 0
for index, row in matches_processed.iterrows():
    if(row['toss_winner'] == row['team1']):
        toss_loss_count[row['team2']] = toss_loss_count[row['team2']]+1
    elif(row['toss_winner'] == row['team2']):
        toss_loss_count[row['team1']] = toss_loss_count[row['team1']]+1
toss_loss_count
toss_loss_countdf = pd.DataFrame.from_dict(toss_loss_count, orient = 'index', columns = ['Matches'])

toss_loser_match_winner = matches_processed[(matches_processed['toss_winner']!=matches_processed['winner'])\
                                           &(matches_processed['winner']!='N/A')]
match_winner_count = toss_loser_match_winner[['id','winner']].groupby('winner').count()

toss_loss_countdf.reset_index(inplace=True)
match_winner_count.reset_index(inplace=True)

match_winner_count.rename(columns={'winner':'Teams', 'id':'Matches'},inplace=True)
toss_loss_countdf.rename(columns={'index':'Teams'},inplace=True)

match_winner_count['param'] = 'Toss Loss Match Wins'
toss_loss_countdf['param'] = 'Total Toss Loss'

frames = [match_winner_count, toss_loss_countdf]
result = pd.concat(frames)

%matplotlib inline
g = sns.catplot(x='Teams', y='Matches', hue='param', data=result, kind='bar', aspect=2.5)
g.set_xticklabels(rotation=90)
g.fig.suptitle("toss loss but match win trend")
```

- Overall __Royal Challenger Bangalore__ lost maximum _toss_.
- __Mumbai Indians__ managed to emarged as victorious maximum times even after losing the toss.
